# **By using the above method(info) and observing the dataset with head method it seems that some of the columns have integer and float values but the datatype of that particular column is object. So to do a proper analysis we need to change the datatype of those particular column into Float/Int datatype**

```python
df=df.astype({'Reviews':'float'})
```

```python
df.info()
```

```python
df['Price'] = df['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x))
```

```python
df=df.astype({'Price':'float'})
```

```python
df['Installs']
```

```python
df['Installs']=df['Installs'].apply(lambda x: str(x).replace('+','') if '+' in str(x) else str(x))
```

```python
df['Installs']=df['Installs'].apply(lambda x: str(x).replace(',','')if ',' in str(x) else str(x))
```

```python
df=df.astype({'Installs':'float'})
```

```python
df.info()
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
```

```python
def convert_to_mb(column):
  if "M" in column:
    result = float(column.split('M')[0])
  elif "k" in column:
    result = float(column.split("k")[0]) * 0.001
  elif column == "Varies with device":
    result = np.NaN
  return result
```

```python
df["Size_in_MB"]=df.Size.apply(convert_to_mb)
df["Size_in_MB"].fillna(value=df.Size_in_MB.mean(), inplace=True)
```

```python
#Plotting a scatter plot between sentiment polarity and sentiment subjectivity.
plt.figure(figsize=(15,8))
plt.scatter(merged_df['Sentiment_Polarity'],merged_df['Sentiment_Subjectivity'])
plt.xlabel('Sentiment_Polarity')
plt.title('Sentiment analysis of reviews')
```

Scatter plots' primary uses are to observe and show relationships between two numeric variables. The dots in a scatter plot not only report the values of individual data points, but also patterns when the data are taken as a whole.

Above figure represent a plot of polarity and subjectivity for all the reviews. While the polarity is concentrated mostly in the center, the subjectivity is spread out across the graph. This indicates that our collection of reviews shows a wide range of subjectivity and most of the reviews fall in [-0.50,0.75] polarity scale implying that the extremely negative or positive sentiments are significantly low.

From the above scatter plot it can be concluded that sentiment subjectivity is not always proportional to sentiment polarity but in maximum number of case, shows a proportional behavior, when variance is too high or low.
